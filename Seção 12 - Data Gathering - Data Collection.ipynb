### O que é Coleta de Dados (Data Collection ou Data Gathering)?

É a etapa em que você transforma algo do mundo (opiniões de clientes, temperatura do ar, cliques em um site, imagens de satélite etc.) em dados estruturados que podem ser analisados.

**Por que a coleta de dados é tão importante?**

**“Garbage in, garbage out”**: 
- (lixo entra, lixo sai): se os dados coletados forem ruins, incompletos ou enviesados, toda análise ou modelo de machine learning posterior será comprometido.
- É geralmente a etapa que mais consome tempo e dinheiro em qualquer projeto de dados (pode chegar a 70–80% do tempo total).

**Tipos de dados mais comuns coletados**

- *Quantitativos*: números, medidas (idade, faturamento, temperatura)
- *Qualitativos*: textos, opiniões, imagens, áudios, vídeos
- *Estruturados*: tabelas, CSV, bancos relacionais
- *Não estruturados*: e-mails, redes sociais, fotos, logs

**Dados podem ter 2 tipos principais:**
- Dados Primários: São dados que você criou 
- Dados Secundários: São dados já existentes que foram coletados organizados por outra pessoa, não por você.

**Como obter dados que estão on-line, mas não estão empacotados em um arquivo para download?**
- Web Scraping
  - É a técnica de extração de informações de sites de forma programática.
  - Em essência, o método se resume em fazer request(solicitação) à página da Web de um servidor e, em seguida, extrair os dados relevantes do documento HTML que é enviado(response/resposta) a você.
- APIs
  - São as "pontes" do mundo digital
  - Foram projetadas especificamente para o intercâmbio de dados programáticos
  - Muito usado para conectar diferentes sistemas
