**Text File (arquivo de texto)**: armazena dados como texto legível (caracteres), usando codificações como UTF-8. Exemplos: `.txt`, `.csv`, `.json`.

**End of Line Character (EOL)**: é o caractere que indica o **fim de uma linha** em um arquivo de texto.
  * No Windows: `\r\n`
  * No Linux/macOS: `\n`

**Binary File (arquivo binário)**: armazena dados em **formato binário** (0s e 1s) não legível diretamente, usado por imagens, vídeos, executáveis etc. Exemplos: `.jpg`, `.exe`, `.mp4`.


**File**:é o arquivo em si, ou seja, o dado armazenado no disco (exemplo: dados.txt).

**File object**: é o objeto criado pelo Python quando você abre um arquivo com a função open(). Ele permite ler, escrever e manipular o conteúdo do arquivo.

**Reading (leitura)**: é o ato de ler o conteúdo bruto de um arquivo ou dado (por exemplo, ler todo o texto de um .txt ou os bytes de um .bin).
Exemplo: ler "1,2,3,4" de um arquivo CSV.

**Parsing (análise)**:é o processo de interpretar e estruturar esse conteúdo bruto, transformando-o em algo compreensível pelo programa.
Exemplo: converter "1,2,3,4" em uma lista [1, 2, 3, 4].
